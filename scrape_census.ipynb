## Pull retail wages from census

```python
import requests
import json

def get_census_data(api_key, state_code, county_code):
    base_url = f"https://api.census.gov/data/2021/acs/acs5?get=NAME,B24031_001E&for=tract:*&in=state:{state_code}+county:{county_code}&key="
    url = f"{base_url}{api_key}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Request failed with status code: {response.status_code}")
        return None
```

```python
# get all state and county fips codes and loop over them
fn = "/Users/davidvandijcke/Dropbox (University of Michigan)/github/census_wrangle/data/in/safegraph_open_census_data_2020/metadata/cbg_fips_codes.csv"
fips = pd.read_csv(fn)

fips['county_fips'] = fips['county_fips'].astype(str).str.zfill(3)
fips['state_fips'] = fips['state_fips'].astype(str).str.zfill(2)

fips = fips.drop_duplicates(['county_fips', 'state_fips'])
```

```python
fn = os.path.join(dir_data, "wages", "wages_B24031_001_2021.csv")
for index, row in fips.iterrows():
    state_code = row['state_fips']
    county_code = row['county_fips']
    census_data = get_census_data(api_key, state_code, county_code)
    if census_data:
        header = census_data.pop(0)
        df = pd.DataFrame(census_data, columns=header)
        df.drop(columns=['NAME'], inplace=True)
        df.to_csv(fn, mode= "a",  index=False,
                  header=not os.path.exists(fn))
```
